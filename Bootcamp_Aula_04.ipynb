##Desafio: adicionar uma coluna mostrando a região de cada Estado.

###Dica: usaveis.sum(axis=1) sendo o axis para mudar o eixo.

```python
regiao =  {'1':'Norte',
           '2':'Nordeste',
           '3':'Sudeste',
           '4':'Sul',
           '5':'Centro-Oeste'}
function = lambda x: regiao[x.name[0]]
usaveis['Região'] = usaveis.apply(function, axis=1)
```
